```python
import torch
import torch.nn as nn
import numpy as np
```

```python
class NN(nn.Module):

# init
    def __init__(self):

        super(Net, self).__init__()

        self.fc1 = nn.Linear(32**2, 16**2)

        self.fc2 = nn.Linear(16**2, 4**2)

        self.fc3 = nn.Linear(4 **2, 1)

# forward prop

    def forward(self, x):

        sigmoid = nn.Sigmoid()

        x = sigmoid(self.fc1(x))
        x = sigmoid(self.fc2(x))
        x = sigmoid(self.fc3(x))

        return x
```

```python
input_layer = torch.rand(32**2) # input data

%%time
net = NN()
result = net(input_layer) # start forward

print(result)
```

```python
arr = [net.fc1.weight.flatten(), net.fc1.bias, 
       net.fc2.weight.flatten(), net.fc2.bias, 
       net.fc3.weight.flatten(), net.fc3.bias]

for i in range(len(arr)):
    arr[i] = arr[i].detach().numpy()
    
res = np.concatenate(arr)
```

```python
np.save('weight', res)
np.save('input', input_layer.numpy())
```
